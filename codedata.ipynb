```python
import pyarrow.parquet as pq
import random

file_path = "code_corpus.parquet"

print(f"✓ 파일: {file_path}\n")

# ParquetFile 객체 생성
parquet_file = pq.ParquetFile(file_path)

total_rows = parquet_file.metadata.num_rows
total_row_groups = parquet_file.num_row_groups

print(f"총 행 수: {total_rows:,}")
print(f"총 Row Group 수: {total_row_groups}")

# 랜덤으로 row group 선택
random_row_groups = random.sample(range(total_row_groups), min(5, total_row_groups))

print(f"\n랜덤 선택된 Row Group: {random_row_groups}")
print("\n" + "=" * 70)
print("🎲 랜덤 샘플 10개")
print("=" * 70)

samples_collected = []

# 선택된 row group에서 데이터 읽기
for rg_idx in random_row_groups:
    if len(samples_collected) >= 10:
        break
    
    try:
        table = parquet_file.read_row_group(rg_idx)
        
        # 이 row group에서 랜덤하게 선택
        num_rows = len(table)
        num_to_sample = min(10 - len(samples_collected), num_rows)
        
        if num_rows > 0:
            random_indices = random.sample(range(num_rows), num_to_sample)
            
            for idx in random_indices:
                sample = {}
                for col_name in table.column_names:
                    sample[col_name] = table.column(col_name)[idx].as_py()
                samples_collected.append(sample)
    except:
        continue

# 10개 출력
for i, sample in enumerate(samples_collected[:10], 1):
    print(f"\n{'='*70}")
    print(f"샘플 #{i}")
    print(f"{'='*70}")
    
    for key, value in sample.items():
        if value is None:
            print(f"\n[{key}]: (null)")
        elif isinstance(value, str):
            if key == 'text':
                # 코드는 전체 출력
                print(f"\n[{key}]:")
                print("-" * 60)
                print(value)
                print("-" * 60)
            elif len(value) > 100:
                print(f"\n[{key}]: {value[:100]}...")
            else:
                print(f"\n[{key}]: {value}")
        else:
            print(f"\n[{key}]: {value}")

print("\n" + "=" * 70)
print("✅ 완료!")
print("=" * 70)
```
